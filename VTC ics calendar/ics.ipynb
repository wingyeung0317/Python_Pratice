```python
import pandas as pd
from urllib.request import urlopen
import datetime as dt
import ics
import re
import arrow
```

```python
url = 'https://moodle2324.vtc.edu.hk/calendar/export_execute.php?userid=85982&authtoken=93b69ceb7e9a9ee7e76b48f0e6878457ebc12196&preset_what=all'
```

```python
icsFile = ics.Calendar(urlopen(url).read().decode('utf-8'))
# icsFile = ics.Calendar(open('icalexport.ics').read()) # Test Calendar
```

```python
icsdf = pd.DataFrame()
for event in icsFile.events:
    eventDate = event.end.to('Asia/Hong_Kong')
    icsdf = pd.concat([icsdf, pd.DataFrame({'name':[event.name], \
                                            'course':[event.categories], \
                                            'description':[event.description], \
                                            'due':[eventDate], \
                                            'url':[f"https://moodle2324.vtc.edu.hk/calendar/view.php?view=day&time={int(eventDate.timestamp())}"]})])
```

```python
submissiondf = icsdf[icsdf['name'].str.contains('is due')]
submissiondf['name'] = submissiondf['name'].apply(lambda x: re.sub(' is due', '', x))
submissiondf['course'] = submissiondf['course'].apply(lambda x: list(x)[0])
submissiondf = submissiondf.reset_index(drop=1)
submissiondf
```
